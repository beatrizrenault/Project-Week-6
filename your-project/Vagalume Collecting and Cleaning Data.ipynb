## Get rank 2019 from Google

```python
url = 'https://www.google.com/search?q=musicas+mais+tocadas+brasil+2019&oq=musicas+mais+tocadas+brasil+2019&aqs=chrome..69i57j0l7.8493j0j7&sourceid=chrome&ie=UTF-8'
html = requests.get(url).content
soup = BeautifulSoup(html, "lxml")

song = soup.find_all('div',attrs = {'class':'BNeawe s3v9rd AP7Wnd'}) [:51]
song_2019 = [item.get_text() for item in song]

singer = soup.find_all('div',attrs = {'class':'BNeawe tAd8D AP7Wnd'} ) [1:52]
singer_2019 = [item.get_text().replace(' · 2019','').replace('&','e') for item in singer]

print(song_2019)
print(singer_2019)
```

```python
info_2019 = []

for singer, song in zip(singer_2019, song_2019):
        info_2019.append((singer,song))

```

```python
#fixing SONG SPELLING error manually:

song_2019[song_2019.index('Bebi Minha Bicicleta (Coração Falido)')] = 'Bebi a Bicicleta'

#fixing SINGER NAMING error manually:
singer_2019[song_2019.index('Na Cama Que eu Paguei')] = 'Wesley Safadão'
singer_2019[song_2019.index('3 Batidas')] = 'Guilherme e Benuto'
singer_2019[song_2019.index('Quarta Cadeira')] = 'Matheus e Kauan'
singer_2019[song_2019.index('Casal Raiz')] = 'Xand Avião'
singer_2019[song_2019.index('SENTADÃO')] = 'Pedro Sampaio'
singer_2019[song_2019.index('Combatchy')] = 'Anitta'
singer_2019[song_2019.index('Invocada')] = 'Ludmilla'
singer_2019[song_2019.index('My Only One')] = 'Sebastian Yatra'
singer_2019[song_2019.index('Meu mel')] = 'Anitta'
singer_2019[song_2019.index('Verdinha')] = 'Ludmilla'
```

```python
info_2019 = []
for singer, song in zip(singer_2019, song_2019):
        info_2019.append((singer,song))
info_2019

df2019 = pd.DataFrame(info_2019, columns =['Singer', 'Song'])
df2019.head(5)
```
